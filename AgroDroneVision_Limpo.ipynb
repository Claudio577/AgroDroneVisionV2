```python
# ===================================================
# 🌾 AgroDrone Vision — NDVI Analysis (Sentinel-2)
# ===================================================
# Autor: Claudio Yoshida
# Projeto: edufin-ai-cloud
# Data: 20/10/2025
# Descrição: Análise de vegetação via NDVI usando Earth Engine
# ===================================================

# 1️⃣ Instalar dependências
!pip install earthengine-api folium geemap rasterio geopandas -q

# 2️⃣ Importar bibliotecas
import ee, folium, geemap

# 3️⃣ Autenticar e inicializar o Google Earth Engine
ee.Authenticate()  # Execute e siga o link
ee.Initialize(project='edufin-ai-cloud')
print("✅ Earth Engine conectado com sucesso!")

# 4️⃣ Definir área de interesse (exemplo: Paraná)
area = ee.Geometry.Rectangle([-51.3, -23.6, -51.2, -23.5])

# 5️⃣ Buscar imagem Sentinel-2 com pouca nuvem
img = (ee.ImageCollection('COPERNICUS/S2_SR')
       .filterBounds(area)
       .filterDate('2025-08-01', '2025-08-10')
       .sort('CLOUD_COVER')
       .first())

# 6️⃣ Calcular NDVI
ndvi = img.normalizedDifference(['B8', 'B4']).rename('NDVI')

# 7️⃣ Exibir mapa interativo
Map = geemap.Map(center=[-23.55, -51.25], zoom=13)
Map.addLayer(ndvi, {'min': 0, 'max': 1, 'palette': ['red','yellow','green']}, 'NDVI')
Map.addLayerControl()
Map
```
